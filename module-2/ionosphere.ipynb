```python
import matplotlib.figure
from sklearn.metrics import roc_curve, precision_recall_curve, auc, confusion_matrix
import matplotlib.pyplot as plt
import matplotlib.cm
import numpy as np
import seaborn as sns
import matplotlib
from itertools import product
from typing import List

def plot_curve(x_values, y_values, auc, thresholds, title, x_label, y_label, random_guessing_line=None) -> matplotlib.figure.Figure:
    fig, ax_main = plt.subplots(figsize=(8, 6))
    ax_main.plot(x_values, y_values, label='Curve (AUC = %0.2f)' % auc)
    ax_main.set_xlim([0.0, 1.0])
    ax_main.set_ylim([0.0, 1.05])
    ax_main.set_xlabel(x_label)
    ax_main.set_ylabel(y_label)
    ax_main.set_title(title)
    fig.legend(loc="lower right")
    
    if random_guessing_line is not None:
        random_guessing_line_start = random_guessing_line[0]
        random_guessing_line_end = random_guessing_line[1]
        ax_main.plot(random_guessing_line_start, random_guessing_line_end, 'k--', label="Random guessing line")  # Random guessing line

    # first filter out infinite thresholds by replacing them with the max or min
    max_t = thresholds[~np.isinf(thresholds)].max()
    min_t = thresholds[~np.isinf(thresholds)].min()
    thresholds = [max_t if np.isposinf(t) else t for t in thresholds]
    thresholds = [min_t if np.isneginf(t) else t for t in thresholds]

    ax_threshold = ax_main.twinx()
    ax_main.plot(x_values[:len(thresholds)], thresholds, markeredgecolor='r', linestyle='dashed', color='r')
    ax_threshold.set_ylabel('Threshold', color='r')

    fig.show()
    return fig


def plot_confusion_matrix(
    cm, labels, title, color="Blues"
) -> matplotlib.figure.Figure:
    fig = plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap=color)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.xticks([0.5 + i for i in range(len(labels))], labels)
    plt.yticks([0.5 + i for i in range(len(labels))], labels)
    return fig
```

```python
from torch.utils.data import DataLoader, TensorDataset
import torch
```

```python
DEVICE = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")
DEVICE = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
DEVICE
```

```python
X_train = torch.tensor(X_train_df.values, dtype=torch.float32).to(DEVICE)
y_train = torch.tensor(y_train_df.values, dtype=torch.float32).to(DEVICE)
X_test = torch.tensor(X_test_df.values, dtype=torch.float32).to(DEVICE)
y_test = torch.tensor(y_test_df.values, dtype=torch.float32).to(DEVICE)

(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from torch.nn import Sequential, Linear, ReLU, Sigmoid
```

```python
model = Sequential(
    Linear(X_train.shape[1], 16),
    ReLU(),
    Linear(16, 16),
    ReLU(),
    Linear(16, 1),
    Sigmoid()
).to(DEVICE)

model
```

```python
train_ds = TensorDataset(X_train, y_train)
test_ds = TensorDataset(X_test, y_test)

train_dl = DataLoader(train_ds, batch_size=32, shuffle=True)
test_dl = DataLoader(test_ds, batch_size=32)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

loss_fn = torch.nn.BCELoss()

num_epochs = 500

early_stopping_patience = 20
```

```python
train_loss_history = []
test_loss_history = []

last_best_loss = float('inf')
epochs_no_improve = 0

for epoch in range(num_epochs):
    model.train()

    avg_train_loss = 0.0
    for xb, yb in train_dl:
        pred = model(xb).squeeze()
        loss = loss_fn(pred, yb)
        avg_train_loss += loss.item()
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
    
    avg_train_loss /= len(train_dl)
    train_loss_history.append(avg_train_loss)
    
    
    model.eval()
    with torch.no_grad():
        avg_test_loss = 0.0
        for xb, yb in test_dl:
            pred = model(xb).squeeze()
            loss = loss_fn(pred, yb)
            avg_test_loss += loss.item()
        avg_test_loss /= len(test_dl)
        test_loss_history.append(avg_test_loss)
    print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_test_loss:.4f}")

    if avg_test_loss < last_best_loss:
        last_best_loss = avg_test_loss
        epochs_no_improve = 0
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        epochs_no_improve += 1

    if epochs_no_improve >= early_stopping_patience:
        print("Early stopping triggered")
        break

model.load_state_dict(torch.load('best_model.pth'))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_loss_history, label='Train Loss')
plt.plot(test_loss_history, label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Test Loss over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def plot_roc_curve(
    y_true, y_score, title="ROC and AUC"
) -> matplotlib.figure.Figure:
    """
    Plot ROC curve for binary classification.
    
    Args:
        y_true: Binary ground truth labels
        y_score: Predicted probability scores
        title: Plot title
    
    Returns:
        Matplotlib figure
    """
    # Check if this is multilabel data
    if isinstance(y_true, np.ndarray) and len(y_true.shape) > 1 and y_true.shape[1] > 1:
        return plot_multilabel_roc_curve(y_true, y_score, title)
        
    # Original implementation for binary data
    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)
    random_guessing_line = ([0, 1], [0, 1])
    return plot_curve(fpr, tpr, roc_auc, thresholds, title, "False Positive Rate", "True Positive Rate", random_guessing_line)
```

```python
def plot_histogram_probabilities(
    y_score,
    y_true,
    title: str,
    classes: list[str] = ['Positive', 'Negative'],
) -> matplotlib.figure.Figure:
    fig, ax = plt.subplots()
    ax.hist(y_score[y_true == 1], bins=21, color='green', alpha=0.5)
    ax.hist(y_score[y_true == 0], bins=21, color='red', alpha=0.5)
    ax.set_xlabel('Probability')
    ax.set_ylabel('Count')
    ax.set_xticks([i * 0.05 for i in range(21)])
    plt.xticks(rotation=45)
    ax.legend(classes)
    ax.set_title(title)
    ax.axvline(x=0.5, color='red', linestyle='--')
    return fig
```
